```python
labs_soup=[]
for item in labs:
    soups=[]
    n=1
    while n < 13:
        url = 'https://www.cna.com.tw/cna2018api/api/simplelist/categorycode/'+item[0]+'/pageidx/'+str(n)
        res = requests.get(url)
        soup = BeautifulSoup(res.text,'html5lib')
        body = soup.findAll('body')[0].getText()
        PageUrl = re.findall(r'\"PageUrl\"\:\S[^\,]+',body)
        urls = [txt[11:-1] for txt in PageUrl]
        soups.extend(urls)
        n += 1
        if len(urls) < 20:
            break
    labs_soup.append([soups,item[1]])
```

```python
news={}
for lab in labs_soup:
    lab_name = lab[1]
    lab_urls = lab[0]
    n = 0
    for url in lab_urls:
        res = requests.get(url)
        soup = BeautifulSoup(res.text,'html5lib')
        txt = ""
        try:
            for s in soup.find_all('div',{'class':'paragraph'})[0].findAll('p'):
                txt=txt+s.getText()
        except:IndexError
            
        research = re.search(r'[\）\)\：\:\》](.+[\。\!\?\s])+',txt)
        if research == None or len(research.group())<2:
#        if re.search(r'',txt) == None:
            print(txt)
        else:
            txt = re.search(r'[\）\)\：\:\》](.+[\。\!\?\s])+',txt).group()[1:]


        news[txt]=lab_name
        n += 1
        if n % 20 == 0:
            print((lab_name,n),end=",")
```

```python
import jieba
import nltk
```

```python
stopwords=[]
with open('../30/txt/stopwords_ch.txt', 'r', encoding='UTF-8') as file:
    for data in file.readlines():
        data = data.strip()
        stopwords.append(data)
```

```python
lab_tags ={}
prelab = ''
books = list(news.items())
for book in books:
    for lab in ['生活','運動','政治','科技','國際','兩岸','產經','社會','地方','娛樂','證券','文化']:
        if book[1]==lab:
            if prelab != lab:
                txt=''
            txt = txt+book[0]
            lab_tags[book[1]] = txt
            prelab = lab
```

```python
filterwords=[
    '年','月','中央社','記者','編輯','譯者','日電',
    '\n',' ','「','」'
    ,'今天','說','表示','年','月','日','後','前','時','應','中','今年'
    ,'兩','下','人','最','次','再','先'
]
```

```python
def get_fwords():
    lab_fwords={}
    lab_Ftwowords = {}
    for labname in lab_tags.keys():
        print(labname,end=',')
        lab_words = lab_tags[labname]
    #    seg_list = jieba.cut_for_search(lab_words)#, cut_all=False)
        seg_list = jieba.cut(lab_words, cut_all=False)
        txt = ",".join(seg_list)
        tokens = txt.split(',')
        tokens = list(filter(lambda a: a not in stopwords and a not in filterwords and a.isalpha() == True,tokens))#and len(a)>1, tokens))
        ###
        lab_alltwowords = list(nltk.bigrams(tokens))
        two = nltk.FreqDist(lab_alltwowords).most_common(350)
        lab_Ftwowords[labname] = [(i[0]+i[1],j) for i,j in two]
        ###
        lab_fwords[labname] =  [i for i in nltk.FreqDist(tokens).most_common(350)]
    return lab_fwords,lab_Ftwowords
```

```python
lab_fwords,lab_Ftwowords = get_fwords()
```

```python
#去除交集的高頻字

dic3={}
for lab in lab_fwords.keys():   
    word_content2 = lab_fwords[lab].copy()
    for lab2 in lab_fwords.keys():        
        if lab != lab2:
            word_content = [w for w in lab_fwords[lab] if (w[0] in [a[0] for a in lab_fwords[lab2]]) ]
            for w in word_content:
                try:
                    word_content2.remove(w)                  
                except ValueError:
                    pass
    dic3[lab] = word_content2  
    
    
    
dic2={}
for lab in lab_Ftwowords.keys():   
    word_content2 = lab_Ftwowords[lab].copy()
    for lab2 in lab_Ftwowords.keys():        
        if lab != lab2:
            word_content = [w for w in lab_Ftwowords[lab] if (w[0] in [a[0] for a in lab_Ftwowords[lab2]]) ]
            for w in word_content:
                try:
                    word_content2.remove(w)                   
                except ValueError:
                    pass
    dic2[lab] = word_content2
    
lab_Ftwowords = dic2
lab_fwords = dic3
```

```python
def pos_features(article,top):
    
    ftwowords = [w[0] for v in lab_Ftwowords.values()  for w in v]
    fwords = [w[0] for v in lab_fwords.values()  for w in v]
    
    
    features={}
    t = ''
    seg_list = jieba.cut_for_search(article)
    t = ",".join(seg_list)
    article_tokens = t.split(',')
    article_tokens = [w for w in article_tokens if w.isalpha()]
    
    
    
    
    filt_tokens = list(filter(lambda a: a not in stopwords and a not in filterwords , article_tokens))
    toks = [w[0] for w in nltk.FreqDist(filt_tokens).most_common(top)]
    for tok in toks:
        if tok in fwords:
            features[tok] = True 

    

    twowords = list(nltk.bigrams(article_tokens))
    alltwowords2 = [i[0]+i[1] for i in twowords]
    intwowords = [two for two in alltwowords2 if two in All_Ftwowords]
    ftwowords = nltk.FreqDist(intwowords).most_common(int(top))
    for i in ftwowords:
        w = i[0]
        if w in ftwowords:
            features[w] = True

            

    return features
```

```python
def get_featuresets(books,top=200):
    n = 0
    featuresets=[]
    for book in books:
        article = book[0]
        labname = book[1]
        features = pos_features(article,top)
        feature_item = (features,labname)
        featuresets.append(feature_item)
        n += 1
        if n % 20 == 0:
            print(n,end=',')

    return featuresets
```

```python
featuresets = get_featuresets(books,200)
```

```python
size = int(len(featuresets)*0.7)
train_set, test_set = featuresets[:size], featuresets[size:]
classifier = nltk.NaiveBayesClassifier.train(train_set)
nltk.classify.accuracy(classifier, test_set)
```
